```python
import plotly.express as px
import pandas as pd

df_clean = pd.read_csv('denuncias_consumidor_2019_2022.csv', sep=';', encoding='utf-8-sig')
# 1. Agrupamos los datos para el gráfico
# Contamos cuántas denuncias hay por Año y por Categoría Agrupada
df_plot = df_clean.groupby(['Año', 'Motivo_Agrupado']).size().reset_index(name='Cantidad')

# 2. Creamos el gráfico de barras con Plotly
fig = px.bar(df_plot, 
             x='Año', 
             y='Cantidad', 
             color='Motivo_Agrupado',
             barmode='group', # Esto pone las barras una al lado de la otra
             title='Evolución de Motivos de Denuncia (2019-2022)',
             labels={'Cantidad': 'Número de Denuncias', 'Motivo_Agrupado': 'Categoría'},
             color_discrete_sequence=px.colors.qualitative.Safe)

fig.update_layout(
    xaxis_type='category', # Para que los años se vean como etiquetas y no como números continuos
    legend_title_text='Categorías de Motivo',
    template='plotly_white'
)

# 4. Mostramos el gráfico
fig.show()

fig.write_html('reporte_motivos_interactivo.html')
```
